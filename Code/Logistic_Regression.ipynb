```python
## Inputting training dataset
movie2 = pd.read_csv("/Users/jenryyu/Desktop/movie_class.csv")
movie2.head()
```

```python
movie2.dropna(inplace = True)

movie2.shape
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
# Drop the lines with null value

X1 = movie2[['actors_rating']]
X2 = movie2[['directors_rating']]
X3 = movie2[['duration']]
X4 = movie2[['Comedy','Drama','Horror','Musical','Action','Crime','Animation',\
             'Adventure','Biography','Family','Thriller','Fantasy','Romance','Sci-Fi',\
             'Mystery','War','Western','Music','Sport','History']]
y = movie2['G/B']


_ = log_regression.fit(X1, y)
y_pred1 = log_regression.predict(X1)

print('Actors rating')
print('accuracy score:',accuracy_score(y, y_pred1))
print('precision score:',precision_score(y, y_pred1))
print('recall score:',recall_score(y, y_pred1))
print('---------------------------------------------')

_ = log_regression.fit(X2, y)
y_pred2 = log_regression.predict(X2)

print('Directors rating')
print('accuracy score:',accuracy_score(y, y_pred2))
print('precision score:',precision_score(y, y_pred2))
print('recall score:',recall_score(y, y_pred2))
print('---------------------------------------------')

_ = log_regression.fit(X3, y)
y_pred3 = log_regression.predict(X3)

print('Duration')
print('accuracy score:',accuracy_score(y, y_pred3))
print('precision score:',precision_score(y, y_pred3))
print('recall score:',recall_score(y, y_pred3))
print('---------------------------------------------')

_ = log_regression.fit(X4, y)
y_pred4 = log_regression.predict(X4)

print('Genre')
print('accuracy score:',accuracy_score(y, y_pred4))
print('precision score:',precision_score(y, y_pred4))
print('recall score:',recall_score(y, y_pred4))
print('---------------------------------------------')


X = movie2[['month_published','duration','directors_rating','actors_rating',\
             'Comedy','Drama','Horror','Musical','Action','Crime','Animation',\
             'Adventure','Biography','Family','Thriller','Fantasy','Romance','Sci-Fi',\
             'Mystery','War','Western','Music','Sport','History']]
y = movie2['G/B']

_ = log_regression.fit(X, y)
y_pred = log_regression.predict(X)

print('Comprehensive data')
print('accuracy score:',accuracy_score(y, y_pred))
print('precision score:',precision_score(y, y_pred))
print('recall score:',recall_score(y, y_pred))
```
